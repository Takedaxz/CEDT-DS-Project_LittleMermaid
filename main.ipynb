```python
import json
import pandas as pd

file_path = '/Users/pookansmacbookpro/Documents/CEDT Computer Engineering/Sem1_2024/2110403_DataSci/Project/Data_2018-2023/2018/201800002'

import pandas as pd
import json

# Load JSON data (Replace 'file_path.json' with your actual file path)
with open(file_path, 'r') as f:
    data = json.load(f)

# Extract relevant information from the JSON structure
records = []

item = data.get('abstracts-retrieval-response', {}).get('item', {})

# Extract relevant fields with default values for missing data
title = item.get('bibrecord', {}).get('head', {}).get('citation-title', '')
abstract = item.get('bibrecord', {}).get('head', {}).get('abstracts', '')

authors = []
for author in data.get('abstracts-retrieval-response', {}).get('authors',{}).get('author',[]):
    name=author.get('preferred-name', {}).get('ce:indexed-name', '')
    if name:
        authors.append(name)

# Initialize list to hold all classification codes
classification_codes = []
# Access classifications within enhancement > classificationgroup
for classification_group in item.get('bibrecord', {}).get('head', {}).get('enhancement', {}).get('classificationgroup', {}).get('classifications', []):
    classifications = classification_group.get('classification', [])
    
    if isinstance(classifications, list):
        for classification in classifications:
            # If classification is a dictionary, get classification-code
            if isinstance(classification, dict):
                code = classification.get('classification-code')
                if code:
                    classification_codes.append(code)
    # If classifications is a single dictionary (e.g., for FLXCLASS)
    elif isinstance(classifications, dict):
        code = classifications.get('classification-code')
        if code:
            classification_codes.append(code)

# Extract date of publication
publication_date = data.get('abstracts-retrieval-response', {}).get('coredata', {}).get('prism:coverDate', {})

# Process author affiliations
affiliations = set()
for author_group in item.get('bibrecord', {}).get('head', {}).get('author-group', []):
    affiliation_id = author_group.get('affiliation', {}).get('@afid', {})
    affiliations.add(affiliation_id)

# Extract citations/references
references = []
for reference in item.get('bibrecord', {}).get('tail', {}).get('bibliography', {}).get('reference', []):
    ref_title = reference.get('ref-info', {}).get('ref-title', {}).get('ref-titletext', '')
    if ref_title:
        references.append(ref_title)

# Extract keywords
keywords = [kw.get('$', '') for kw in item.get('citation-info', {}).get('author-keywords', {}).get('author-keyword', [])]

# Store each extracted record in the records list
records.append({
    'Title': title,
    'Abstract': abstract,
    'Author': "; ".join(authors),
    'Classification_Codes': "; ".join(classification_codes),
    'Publication_Date': publication_date,
    'Affiliations_ID': "; ".join(affiliations),
    'References': "; ".join(references),
    'Keywords': "; ".join(keywords)
})

for record in records:
    print(f"Title : {record['Title']}")
    print(f"Abstract : {record['Abstract']}")
    print(f"Author : {record['Author']}")
    print(f"Classification Codes : {record['Classification_Codes']}")
    print(f"Publication Date : {record['Publication_Date']}")
    print(f"Affiliations ID: {record['Affiliations_ID']}")
    print(f"References : {record['References']}")
    print(f"Keywords : {record['Keywords']}")
    print("\n")
# # Convert list of records to DataFrame
# df = pd.DataFrame(records)

# # Save DataFrame to CSV
# # df.to_csv('output.csv', index=False)
# df.head()
```
